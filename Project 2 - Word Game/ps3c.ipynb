```python
import random
import matplotlib.pyplot as plt
#%matplotlib inline

import ps2
import ps3a
```

```python
def plot_mc_hand(hand, N, n):
    
    _, mc_scores = ps3a.play_n_mc_hand(hand, N, n)
    title = print("Histogram of ", N, "Monte Carlo samples with ", n," times playing a hand")
    plot = plt.hist(mc_scores), plt.title(title)
    
    return plot
    # IMPLEMENT ME
```

```python
hand = ps2.get_frequency_dict('helloworld')
random.seed(1)
for n in (10, 100):
    for N in (1, 10, 100):
        x = plot_mc_hand(hand, N, n)
        plt.show(x)
        
        
```
